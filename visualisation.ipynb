```python
# bar chart
#categorical to numerical- bar chart

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=np.array(['Jan','Feb','March','April'])
y=np.array([300,200,500,300])
plt.bar(x,y,color='g',width=0.4)
plt.title('Bar chart')
plt.show()
```

```python
data=[30,50,70,60,40,20,10,98, 67]
plt.hist(data,bins=20,edgecolor='black')
plt.show()
```

```python
# Subplot
# Create multiple plots
plt.figure(figsize=(10,5))
plt.subplots_adjust(hspace=0.5,wspace=0.8) # hspace-space between rows, wspace=columns

# 4 plot 2*2
plt.subplot(2,2,1)
plt.bar(x,y,color='g',width=0.4)
plt.title('Bar chart1')

plt.subplot(2,2,2)
plt.bar(x,y,color='g',width=0.4)
plt.title('Bar chart2')

plt.subplot(2,2,3)
plt.hist(data,bins=20,edgecolor='black')

plt.subplot(2,2,4)
plt.hist(data,bins=20,edgecolor='black')

plt.savefig('subplottt.png')
plt.show()
```

```python
def count_hobby_words(hobbies,word):
    counter = 0
    for hobby in hobbies:
        if word in hobby.split():
            counter +=1
    return counter
```
